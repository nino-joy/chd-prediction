```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import keras
from sklearn.metrics import balanced_accuracy_score
from sklearn import metrics
```

```python
df=pd.read_csv('frs.csv')
df=df.drop(columns=['Unnamed: 0'])
df=df.drop(columns=['frs'])

df.head()

```

```python
X = df.drop(columns=['Target'])

Y = df[['Target']]
X.shape

```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = keras.models.Sequential()


model.add(Dense(12, activation='relu', input_shape=(9,)))

model.add(Dense(71, activation='relu'))
model.add(Dense(51, activation='relu'))
model.add(Dense(31, activation='relu'))
model.add(Dense(11, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train,validation_split=0.1, epochs=1000,batch_size=32, callbacks=[keras.callbacks.EarlyStopping(monitor='val_loss',patience=200)])
```

```python
# accuracy on training data
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
Y_predict  = model.predict(X_test)

Y_pred =np.round(Y_predict ,0)
confusion_matrix(Y_test,Y_pred )

#confusion_matrix(Y_pred ,Y_test)
```

```python
accuracy_score(Y_test,Y_pred )
#from sklearn.metrics import r2_score
#r2_score(Y_test,Y_predict)
```

```python
import sklearn
sklearn.metrics.precision_recall_fscore_support(Y_test, Y_pred, average='micro')
```
